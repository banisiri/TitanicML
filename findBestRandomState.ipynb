```python
y=data_train["Survived"]

features = ["PassengerId","Pclass", "Sex", "SibSp", "Parch"]
data_train = pd.get_dummies(data_train[features])
data_test = pd.get_dummies(data_test[features])
```

```python
#y=data_train["Survived"]
#data_train.drop(['Survived'], axis=1, inplace=True)
```

```python
from sklearn.ensemble import RandomForestClassifier

best_accuracy = 0
best_random_state = None

for random_state in range(1, 101):
    model = RandomForestClassifier(random_state=random_state)
    model.fit(data_train, y)
    accuracy = model.score(data_train, y)

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_random_state = random_state

print(f"En iyi random state değeri: {best_random_state}")
```

```python
best_accuracy = 0
best_n_estimators = None

for n_estimators in range(1, 101):
    model = RandomForestClassifier(n_estimators=n_estimators, random_state=42)
    model.fit(data_train, y)
    accuracy = model.score(data_train, y)

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_n_estimators = n_estimators

print(f"En iyi n_estimators değeri: {best_n_estimators}")
```

```python
best_accuracy = 0
best_max_depth = None

for max_depth in range(1, 21):
    model = RandomForestClassifier(max_depth=max_depth, random_state=42)
    model.fit(data_train, y)
    accuracy = model.score(data_train, y)

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_max_depth = max_depth

print(f"En iyi max_depth değeri: {best_max_depth}")
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=49, max_depth=20, random_state=1)
rf_model.fit(data_train, y)

TrainAccuracy=rf_model.score(data_train, y)
print(f"Accuracy:{TrainAccuracy:.4f}")
```

```python
id=data_test["PassengerId"]
```

```python
predicts=rf_model.predict(data_test)
df_submit=pd.DataFrame({"PassengerId":id,"Survived":predicts})
df_submit.to_csv("sub5.csv",index=False)
```
