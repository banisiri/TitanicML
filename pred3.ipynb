```python
import pandas as pd
import numpy as np
```

```python
gender_sub=pd.read_csv("gender_submission.csv")
#gender_sub.head()
```

```python
data_train=pd.read_csv("train.csv")
#data_train.head()
```

```python
data_test=pd.read_csv("test.csv")
#data_test

id=data_test["PassengerId"]
```

```python
data_train["Age"].fillna(data_train["Age"].median(), inplace=True)
data_train["Fare"].fillna(data_train["Fare"].median(), inplace=True)

data_test["Age"].fillna(data_test["Age"].median(), inplace=True)
data_test["Fare"].fillna(data_test["Fare"].median(), inplace=True)
```

```python
data_train.drop(["Name","Ticket","Cabin"], axis=1, inplace=True)
data_test.drop(["Name","Ticket","Cabin"], axis=1, inplace=True)
```

```python
data_train.dropna(axis=0, inplace=True)
```

```python
data_train.isnull().sum(), data_test.isnull().sum()
```

```python
X = data_train[["Sex", "Embarked"]]
Y = data_test[["Sex", "Embarked"]]

typesTrain=pd.get_dummies(X, drop_first=False)
typesTest=pd.get_dummies(Y, drop_first=False)
```

```python
typesTrain.head()
```

```python
typesTest.head()
```

```python
mapping_boolean = {True: 1, False: 0}
typesTrain['Sex_female'] = typesTrain['Sex_female'].map(mapping_boolean)
typesTrain['Sex_male'] = typesTrain['Sex_male'].map(mapping_boolean)
typesTrain['Embarked_C'] = typesTrain['Embarked_C'].map(mapping_boolean)
typesTrain['Embarked_Q'] = typesTrain['Embarked_Q'].map(mapping_boolean)
typesTrain['Embarked_S'] = typesTrain['Embarked_S'].map(mapping_boolean)

mapping_boolean = {True: 1, False: 0}
typesTest['Sex_female'] = typesTest['Sex_female'].map(mapping_boolean)
typesTest['Sex_male'] = typesTest['Sex_male'].map(mapping_boolean)
typesTest['Embarked_C'] = typesTest['Embarked_C'].map(mapping_boolean)
typesTest['Embarked_Q'] = typesTest['Embarked_Q'].map(mapping_boolean)
typesTest['Embarked_S'] = typesTest['Embarked_S'].map(mapping_boolean)
```

```python
typesTrain.head()
```

```python
typesTest.head()
```

```python
data_train=pd.concat([data_train, typesTrain], axis=1)
data_test=pd.concat([data_test, typesTest], axis=1)

id=data_test["PassengerId"]
```

```python
data_train.head()
```

```python
data_test.head()
```

```python
data_train.drop(["PassengerId","Sex","Embarked"], axis=1, inplace=True)
data_test.drop(["PassengerId","Sex","Embarked"], axis=1, inplace=True)
```

```python
data_test.head()
```

```python
data_train.head()
```

```python
data_train.isnull().sum()
```

```python
data_test.isnull().sum()
```

```python
y=data_train["Survived"]
data_train.drop(['Survived'], axis=1, inplace=True)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=1)
rf_model.fit(data_train, y)

TrainAccuracy=rf_model.score(data_train, y)
print(f"Accuracy:{TrainAccuracy:.4f}")
```

```python
from catboost import CatBoostClassifier

catboost_model = CatBoostClassifier(iterations=100, random_state=1)
catboost_model.fit(data_train, y)

TrainAccuracy = catboost_model.score(data_train, y)
print(f"Accuracy: {TrainAccuracy:.4f}")
```

```python
predicts=catboost_model.predict(data_test)
df_submit=pd.DataFrame({"PassengerId":id,"Survived":predicts})
df_submit.to_csv("sub8.csv",index=False)
```
